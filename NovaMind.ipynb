```python
from datasets import Dataset

def prepare_datasets(tokenizer, data_list):

    full_dataset = Dataset.from_list(data_list)

    split_dataset = full_dataset.train_test_split(test_size=0.2, seed=42)

    def preprocess_function(examples):
        combined_texts = [
            q + "\n" + a + tokenizer.eos_token
            for q, a in zip(examples["question"], examples["answer"])
        ]
        model_inputs = tokenizer(
            combined_texts,
            truncation=True,
            padding="max_length",
            max_length=128
        )

        model_inputs["labels"] = model_inputs["input_ids"].copy()
        return model_inputs

    tokenized_datasets = split_dataset.map(preprocess_function, batched=True)

    return split_dataset, tokenized_datasets

print("Data processing function ready.")
```

```python
def compute_generation_metrics(model, tokenizer, test_dataset):
    """
    Generates responses for the test set and calculates BLEU/ROUGE.
    """
    bleu = evaluate.load("bleu")
    rouge = evaluate.load("rouge")

    predictions = []
    references = []


    model.eval()

    for item in test_dataset:

        prompt = item['question'] + "\n"
        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

        # Generate
        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=100,
                pad_token_id=tokenizer.eos_token_id
            )

        # Decode
        generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Strip the prompt to isolate the answer
        answer_only = generated_text.replace(prompt, "").strip()

        predictions.append(answer_only)
        references.append(item['answer'])

    # Compute Scores
    bleu_score = bleu.compute(predictions=predictions, references=references)
    rouge_score = rouge.compute(predictions=predictions, references=references)

    return {
        "bleu": bleu_score['bleu'],
        "rouge1": rouge_score['rouge1'],
        "rouge2": rouge_score['rouge2'],
        "rougeL": rouge_score['rougeL']
    }
```

```python
# Experiment Configuration
lr_schedules = ["constant"]

experiments = [
    # {"name": "baseline_r8", "r": 8, "lora_alpha": 16, "lr": 2e-4, "epochs": 10},
    # {"name": "high_rank_r16", "r": 16, "lora_alpha": 32, "lr": 2e-4, "epochs": 10},
    {"name": "rank_32", "r": 32, "lora_alpha": 64, "lr": 2e-4, "epochs": 10},
    # {"name": "best_lr_test", "r": 32, "lora_alpha": 64, "lr": 1e-4, "epochs": 10},
]

model_name = "Qwen/Qwen2.5-0.5B"
for lr_schedule_type in lr_schedules:
    for config in experiments:
        print("=="*60)
        print(f"\n=== Starting Experiment: {config['name']} ===")

        # Cleanup
        if 'model' in globals(): del model
        if 'trainer' in globals(): del trainer
        torch.cuda.empty_cache()
        gc.collect()

        # Init WandB
        run = wandb.init(project="NovaMind", name=lr_schedule_type+"_"+config['name'], config=config, reinit=True)

        # Load Resources
        tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True, add_eos_token=True)
        if tokenizer.pad_token is None: tokenizer.pad_token = tokenizer.eos_token

        raw_datasets, tokenized_datasets = prepare_datasets(tokenizer, dataset_content)

        model = AutoModelForCausalLM.from_pretrained(model_name, trust_remote_code=True)

        # LoRA Config
        peft_config = LoraConfig(
            r=config['r'],
            lora_alpha=config['lora_alpha'],
            target_modules=["q_proj", "k_proj", "v_proj", "o_proj", "up_proj", "down_proj"],
            task_type=TaskType.CAUSAL_LM,
            # task_type = TaskType.QUESTION_ANS,
            lora_dropout=0.05
        )
        model = get_peft_model(model, peft_config)

        # Training Args
        args = TrainingArguments(
            output_dir=f"./results/{config['name']}",
            per_device_train_batch_size=4,
            gradient_accumulation_steps=4,
            num_train_epochs=config["epochs"],
            learning_rate=config['lr'],
            logging_steps=1,

            lr_scheduler_type=lr_schedule_type,

            # Evaluation Strategies (Loss)
            eval_strategy="epoch",
            save_strategy="epoch",
            load_best_model_at_end=True,

            report_to="wandb",
            fp16=True
        )

        trainer = Trainer(
            model=model,
            args=args,
            train_dataset=tokenized_datasets['train'],
            eval_dataset=tokenized_datasets['test'],
        )

        trainer.train()

        metrics = compute_generation_metrics(model, tokenizer, raw_datasets['test'])

        # Print answers on test questions
        for question in test_question:
            inputs = tokenizer(question + "\n", return_tensors="pt").to(model.device)
            with torch.no_grad():
                outputs = model.generate(
                    **inputs,
                    max_new_tokens=512,
                    pad_token_id=tokenizer.eos_token_id,
                    repetition_penalty=1.2,  # Penalize exact repetition
                    do_sample=False          # Keep greedy for factual accuracy, or set True for creativity
                )
            generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
            answer_only = generated_text.replace(question + "\n", "").strip()
            print(f"Q: {question}\nA: {answer_only}\n")

        wandb.log(metrics)
        print(f"Metrics for {config['name']}: {metrics}")

        run.finish()
```
